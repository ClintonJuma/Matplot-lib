#Scatter

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
```

```python
y
```

```python
x
```

```python
plt.scatter(x, y, c = "m")
plt.title("Scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Size of scatter dots

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y, s=sizes, c ="g")
plt.title("Size of scatter")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

#Alpha arguments, to change the transparency

```python
plt.scatter(x, y, s=sizes, c ="g", alpha = 0.5)
plt.title("Size of scatter")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

#PIE CHARTS

```python
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.title("PIE CHART")
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.show()
```

#Explode, making the edges stand out

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow ="True")
plt.show()
```

#colors

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["k", "r", "g", "b"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

#legend

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits")
plt.show()
```
